```python
class Dart(object):

    def __init__(self, posi_x=None, posi_y=None, posi_z=None, primary_mass=None, secondary_mass=None):
        self.posi_x = posi_x
        self.posi_y = posi_y
        self.posi_z = posi_z
        self.primary_mass = primary_mass
        self.secondary_mass = secondary_mass
    def distance(self):
        result = np.sqrt(self.posi_x**2 + self.posi_y**2 + self.posi_z**2)
        return result * u.m
    def one(self):
        result = (self.secondary_mass*self.distance())/(self.primary_mass*(1+self.secondary_mass/self.primary_mass))
        return result * u.m
    def bary_x(self):
        result = -(self.secondary_mass*self.posi_x)/(self.primary_mass*(1+self.secondary_mass/self.primary_mass))
        return result * u.m
    def bary_y(self):
        result = -(self.secondary_mass*self.posi_y)/(self.primary_mass*(1+self.secondary_mass/self.primary_mass))
        return result * u.m
    def bary_z(self):
        result = -(self.secondary_mass*self.posi_z)/(self.primary_mass*(1+self.secondary_mass/self.primary_mass))
        return result * u.m
    def two(self):
        result = np.sqrt(self.bary_x()**2 + self.bary_y()**2 + self.bary_z()**2)
        return result * u.m
    
```

```python
primary_random = np.random.randint(730,790,1000) 
secondary_random = np.random.randint(146,182,1000)
primary_randomg = np.random.normal(loc = 760, scale = 30, size = 100000) #gaussian distributed random diameters of the two objects
primary_randomg
secondary_randomg = np.random.normal(loc = 164, scale = 18, size = 100000)
secondary_randomg
total_mass = 5.37e11*u.kg

class Mass(object):                                          #derives the mass of the system from the random diameters
    def __init__(self, p_r=None, s_r=None, total_mass=None):
        self.p_r = p_r
        self.s_r = s_r
        self.total_mass = total_mass 
    def p_mass(self):
        result = (self.total_mass/((self.p_r/2)**3*4/3+(self.s_r/2)**3*4/3))*((self.p_r/2)**3*4/3)
        return result*u.kg
    def s_mass(self):
        result = (self.total_mass/((self.p_r/2)**3*4/3+(self.s_r/2)**3*4/3))*((self.s_r/2)**3*4/3)
        return result*u.kg

p_gauss = Mass(p_r = primary_randomg, s_r = secondary_randomg, total_mass=total_mass) 
pm_gauss = p_gauss.p_mass()
sm_gauss = p_gauss.s_mass()

p_random = Mass(p_r = primary_random, s_r = secondary_random, total_mass=total_mass)
pm_random = p_random.p_mass()
sm_random = p_random.s_mass()

```
